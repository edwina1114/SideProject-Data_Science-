```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
df_preprocessed = pd.read_csv('Absenteeism_preprocessed.csv')
df_preprocessed.describe(include='all')
```

```python
df_preprocessed['Absenteeism Time in Hours'].median()
```

```python
targets = np.where(df_preprocessed['Absenteeism Time in Hours'] > df_preprocessed['Absenteeism Time in Hours'].median(), 
                   1, 0)
targets
```

```python
df_preprocessed['Targets'] = targets
df_preprocessed
```

```python
data_targets = df_preprocessed.drop(['Day of the week', 'Distance to Work', 'Daily Work Load Average', 'Absenteeism Time in Hours'], axis=1)
```

```python
data_targets is df_preprocessed
```

```python
data_targets
```

```python
# Median當作分界線比較穩定，因為他已經先把資料切成大概一半一半
# 45% ~ 55% is acceptable
data_targets['Targets'].sum() / data_targets.shape[0]
```

### Select Features

```python
# data_targets.loc[:, :'Pets']
data_without_targets = data_targets.iloc[:, :-1]
```

```python
unscale_dummies = data_targets.loc[:, 'Reason1':'Reason4']
unscale_dummies.to_numpy
```

```python
unscale_dummies = pd.get_dummies(unscale_dummies, drop_first=True)
unscale_dummies
```

```python
# Dummies no need to scale!
unscale_inputs = data_targets.iloc[:, 4:-1]
unscale_inputs
```

```python
scaler = StandardScaler()
scaled_inputs = scaler.fit_transform(unscale_inputs)
```

```python
scaled_inputs
```

```python
scaled_inputs.shape
```

```python
scaled_with_dummies = np.concatenate((unscale_dummies, scaled_inputs), axis=1)
scaled_with_dummies
```

### Split dataset into train and test then shuffle

```python
# random_state: Shuffle
x_train, x_test, y_train, y_test = train_test_split(scaled_with_dummies, targets, test_size=0.2, random_state=20)
```

```python
print(x_train.shape, y_train.shape)
```

```python
print(x_test.shape, y_test.shape)
```

### Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```python
reg = LogisticRegression()
reg.fit(x_train, y_train)
```

```python
reg.score(x_train, y_train)
```

```python
# Manually Check Accuracy
reg_pred = reg.predict(x_train)
```

```python
reg_pred == y_train
```

```python
np.sum((reg_pred == y_train))
```

### Coefficient, Intercept

```python
feature_names = data_without_targets.columns.values
feature_names
```

```python
summary_table = pd.DataFrame(columns=['Features'], data=feature_names)
summary_table['Coefficient'] = np.transpose(reg.coef_)
summary_table
```

```python
summary_table.index = summary_table.index + 1
summary_table.loc[0] = ['Intercept', reg.intercept_[0]]
summary_table = summary_table.sort_index()
summary_table
```

```python
# A feature is not particularly important:
# if Coefficient is around 0
# if odds ratio is around 1
# -> Daily Work Load Average, Distance to Work, Day of the week, Education
summary_table['Odds_Ratio'] = np.exp(summary_table.Coefficient)
summary_table.sort_values('Odds_Ratio', ascending=False)
```

- Reason1 (Various Diseases): 生病通常也不會去上班，所以Coefficient, Odds_Ratio也是偏高。

- Reason2 (Pregnency, Giving birth): 有些懷孕還是會上班，所以Coefficient, Odds_Ratio會有所下降。

- Reason3 (Poisoning): Coefficient, Odds_Ratio都是最高的（因為中毒一定不會去上班)。

- Reason4 (Light Diseases): 很多人輕微感冒還是會去上班，不太會請假，所以Coefficient, Odds_Ratio會有所下降。

- Daily Work Load Average, Distance to Work有些出乎意料，對於空班率來說影響較小。 -> Drop
